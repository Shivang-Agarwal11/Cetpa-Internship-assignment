```python
try:
    a=int(input("enter 1 number"))
    b=int(input(" enter second number"))
    c=a/b
except:
    print("enter a valid denominator")
else:
    print('result is ',c)
```

```python
try:
    a=int(input("enter 1 number"))
    b=int(input(" enter second number"))
    c=a/b
except ValueError:
    print("enter a integer")
except ZeroDivisionError:
    print("denominator must be non zero")
else:
    print('result is ',c)
```

```python
try:
    a=int(input("enter 1 number"))
    b=int(input(" enter second number"))
    c=a/b
except Exception as t:
    print(t)
else:
    print('result is ',c)
```
